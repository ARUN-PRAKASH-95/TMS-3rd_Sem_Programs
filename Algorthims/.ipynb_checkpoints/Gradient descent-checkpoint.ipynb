```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-10,10,100)
def func(x):
    return x**2
def der(x):
    return 2*x
y = func(x)
gamma = 0.1
x_old = 10
points = np.array([])
for i in range(1000):
    x_new = x_old - der(x_old)*gamma
    points = np.append(points,x_old)
    epsilon = x_new-x_old
    
    #print(epsilon)
    if abs(epsilon)==0:
        break;
    x_old = x_new
plt.plot(x,y);
plt.scatter(points,func(points));
#print(points)
```
